In investing, a time series tracks the movement of stock price over a specified period of time.

# RESAMPLING


Involves changing the frequency of your time series observations
Types: 
1. Down sampling: When you decrease the frequency of the samples.
   Example: Days into months.
2. Up sampling: When you increase the frequency of the samples.
   Examples: Minutes to seconds.

```python
ibm_restructured = ibm.loc[pd.Timestamp('2020-09-01'): pd.Timestamp('2021-05-04')]
plt.plot(ibm_restructured.index, ibm_restructured['Adj Close'])
plt.grid(True)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.xticks(rotation=90)
plt.show()
```

```python
#subplot

f, ax = plt.subplots(2,2, figsize=(10,10), sharex = True, sharey=True)
f.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
f.gca().xaxis.set_major_locator(mdates.MonthLocator())

intel_restructured = intel.loc[pd.Timestamp('2020-09-01'): pd.Timestamp('2021-05-04')]
ax[0,0].plot(intel_restructured.index, intel_restructured['Adj Close'], '.', color='r')
ax[0,0].grid(True)
ax[0,0].tick_params(labelrotation=90)
ax[0,0].set_title('INTEL')


nvda_restructured = nvda.loc[pd.Timestamp('2020-09-01'): pd.Timestamp('2021-05-04')]
ax[0,1].plot(nvda_restructured.index, nvda_restructured['Adj Close'],'.',color='g')
ax[0,1].grid(True)
ax[0,1].tick_params(labelrotation=90)
ax[0,1].set_title('NVDA')


qcom_restructured = qcom.loc[pd.Timestamp('2020-09-01'): pd.Timestamp('2021-05-04')]
ax[1,0].plot(qcom_restructured.index, qcom_restructured['Adj Close'],'.', color='b')
ax[1,0].grid(True)
ax[1,0].tick_params(labelrotation=90)
ax[1,0].set_title('QCOM')


amd_restructured = amd.loc[pd.Timestamp('2020-09-01'): pd.Timestamp('2021-05-04')]
ax[1,1].plot(amd_restructured.index, amd_restructured['Adj Close'], '.', color='y')
ax[1,1].grid(True)
ax[1,1].tick_params(labelrotation=90)
ax[1,1].set_title('AMD')
plt.show()

```

# MOVING WINDOWS

Moving windows are there when you compute the statistics on a window of data represented by a particular period of time and then slide the window across the data by a specified interval i.e., 2 days- 3 days. That way the statistic is continuously calculated as long as the window falss first within the dates of the time series

A rolling mean smothens out short-term fluctuations and highlights longer term trends in data

# daily percentages


```python
daily_close_ibm = ibm[['Adj Close']]

#daily returns
daily_pct_change_ibm = daily_close_ibm.pct_change()

#replace NA values with 0
daily_pct_change_ibm.fillna(0, inplace=True)

daily_pct_change_ibm.head(10)
```

```python
daily_pct_change_ibm.hist(bins=50)
plt.show()
```

# VOLATILITY

The volatility of a stock is a measurement of the change in variance in the returns of a stock over a specific period of time.

To compare the volatility of a stock with another stock to get a feel for which may have less risk or to a market index to examine the stock's volatility in the overall market

Generally, higher the volatility, higher is the risk in investment in that stock

```python
import numpy as np
```

```python
min_periods = 75

#calculate the volatility
#using rolling in order to get the variance
vol = daily_pct_change_ibm.rolling(min_periods).std()*np.sqrt(min_periods)
vol.fillna(0, inplace=True)
vol.tail()
```

```python
vol.plot(figsize=(10,8))
plt.show()
```

# ROLLING MEANS(TRENDS AND SEASONALITY)

```python
ibm_adj_close_px = ibm['Adj Close']
#short moving window rolling mean
ibm['42'] = ibm_adj_close_px.rolling(window=40).mean()

#long moving window rolling mean
ibm['252'] = ibm_adj_close_px.rolling(window=252).mean()

#plot the adjusted closing price, the short and long windows of rolling means
ibm[['Adj Close', '42', '252']].plot(title="IBM")
plt.show()


intel_adj_close_px = intel['Adj Close']
#short moving window rolling mean
intel['42'] = intel_adj_close_px.rolling(window=40).mean()

#long moving window rolling mean
intel['252'] = intel_adj_close_px.rolling(window=252).mean()

#plot the adjusted closing price, the short and long windows of rolling means
intel[['Adj Close', '42', '252']].plot(title="INTEL")
plt.show()


nvda_adj_close_px = nvda['Adj Close']
#short moving window rolling mean
nvda['42'] = nvda_adj_close_px.rolling(window=40).mean()

#long moving window rolling mean
nvda['252'] = nvda_adj_close_px.rolling(window=252).mean()

#plot the adjusted closing price, the short and long windows of rolling means
nvda[['Adj Close', '42', '252']].plot(title="NVDA")
plt.show()



amd_adj_close_px = amd['Adj Close']
#short moving window rolling mean
amd['42'] = amd_adj_close_px.rolling(window=40).mean()

#long moving window rolling mean
amd['252'] = amd_adj_close_px.rolling(window=252).mean()

#plot the adjusted closing price, the short and long windows of rolling means
amd[['Adj Close', '42', '252']].plot(title="AMD")
plt.show()
```
