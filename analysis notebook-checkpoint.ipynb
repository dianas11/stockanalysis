# TIME SERIES DATA


A time series is a sequence of numerical data points taken at successive equally spaced points in time.
In investing, a time series tracks the movement of stock price over a specified period of time.

```python
import matplotlib.pyplot as plt

%matplotlib inline  #makes sure that our plots or graphs are embedded in the notebook
```

```python
import matplotlib.dates as mdates
```

```python
plt.plot(ibm.index, ibm['Adj Close'])
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.grid(True)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.plot(intel.index, intel['Adj Close'])
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.grid(True)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.plot(nvda.index, nvda['Adj Close'])
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.grid(True)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.plot(qcom.index, qcom['Adj Close'])
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.grid(True)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.plot(amd.index, amd['Adj Close'])
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.grid(True)
plt.xticks(rotation=90)
plt.show()
```

```python
#subplots


f, ax = plt.subplots(2,2, figsize=(10,10), sharex = True)
f.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
f.gca().xaxis.set_major_locator(mdates.YearLocator())

ax[0,0].plot(intel.index, intel['Adj Close'], color='r')
ax[0,0].grid(True)
ax[0,0].tick_params(labelrotation=90)
ax[0,0].set_title('INTEL')



ax[0,1].plot(nvda.index, nvda['Adj Close'], color='g')
ax[0,1].grid(True)
ax[0,1].tick_params(labelrotation=90)
ax[0,1].set_title('NVDA')



ax[1,0].plot(qcom.index, qcom['Adj Close'], color='b')
ax[1,0].grid(True)
ax[1,0].tick_params(labelrotation=90)
ax[1,0].set_title('QCOM')



ax[1,1].plot(amd.index, amd['Adj Close'], color='y')
ax[1,1].grid(True)
ax[1,1].tick_params(labelrotation=90)
ax[1,1].set_title('AMD')
plt.show()

```

# RESAMPLING


Involves changing the frequency of your time series observations
Types: 
1. Down sampling: When you decrease the frequency of the samples.
   Example: Days into months.
2. Up sampling: When you increase the frequency of the samples.
   Examples: Minutes to seconds.

```python
ibm_restructured = ibm.loc[pd.Timestamp('2020-09-01'): pd.Timestamp('2021-05-04')]
plt.plot(ibm_restructured.index, ibm_restructured['Adj Close'])
plt.grid(True)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.xticks(rotation=90)
plt.show()
```
